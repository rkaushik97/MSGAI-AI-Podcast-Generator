## Create few shot examples json file

```python
with open('few_shot_examples_response.txt') as file:
  lines = file.readlines()
```

```python
responses = []
response = ""
for line in lines:
    if line.startswith('#'):
        responses.append(response)
        response = ""
        continue
    response += line
```

```python
responses[0]
```

```python
import json

with open('few_shot_examples.json') as file:
  obj = json.load(file)
```

```python
for i, fex in enumerate(obj):
    fex['script'] = responses[i]
    fex['model'] = 'GPT5-mini'
```

```python
with open('few_shot_examples_responses.json', 'w') as file:
    json.dump(obj, file)
```
